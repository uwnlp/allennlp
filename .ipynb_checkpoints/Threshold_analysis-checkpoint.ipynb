```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
snli_hard = pd.read_json('./swabha/snli_test_hard_subset.json', lines=True)
snli_easy = pd.read_json('./swabha/snli_test_easy_subset.json', lines=True)
```

## SNLI hard prediction score histogram

```python
ax = snli_hard.prediction_score.hist(cumulative=False,normed=False,bins=20)
ax.set_ylabel('number of examples')
ax.set_xlabel('score')
```

```python
ax = snli_hard.prediction_score.hist(cumulative=True,normed=True,bins=20)
ax.set_ylabel('% of examples')
ax.set_xlabel('score')
```

## SNLI easy prediction score histogram

```python
ax = snli_easy.prediction_score.hist(cumulative=False,normed=False,bins=20)
ax.set_ylabel('number of examples')
ax.set_xlabel('score')
```

```python
ax = snli_easy.prediction_score.hist(cumulative=True,normed=True,bins=20)
ax.set_ylabel('% of examples')
ax.set_xlabel('score')
```
